```python
import numpy as np

dice_rolls = np.random.randint(1, 7, size=10000)


print(dice_rolls)



```

```python
import numpy as np
import matplotlib.pyplot as plt


values, counts = np.unique(dice_rolls, return_counts=True)

theoretical_count = len(dice_rolls) / 6

plt.figure(figsize=(8, 5))


plt.bar(values, counts,
        width=0.6,
        label='Simulated',
        alpha=0.7,
        color='skyblue',
        edgecolor='black')


plt.axhline(theoretical_count,
            color='red',
            linestyle='--',
            label=f'Theoretical = {theoretical_count:.2f}')

plt.xlabel('Die Face')
plt.ylabel('Frequency')
plt.title('Dice Roll Distribution vs. Theoretical Uniform Distribution')
plt.xticks(values)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

```

```python
'''
What might cause the simulated results to differ from theory?
-Since we're only simulating 10,000 dice rolls there will be some sampling variability, some faces will come up more or less than the ideal 1/6th. 
-The Pseudo Random number generators are deterministic so they can have tiny biases that cause the simulated results to deviate from the uniform distribution.
'''
```
