```python
import numpy as np

dice_rolls = np.random.randint(1, 7, size=10000)


print(len(dice_rolls))



```

```python
import numpy as np
import matplotlib.pyplot as plt


values, counts = np.unique(dice_rolls, return_counts=True)

theoretical_count = [len(dice_rolls) / 6] * 6  # e.g., 10000 / 6 = 1666.67

plt.figure(figsize=(8, 5))
plt.bar(values, counts, width=0.6, label='Simulated', alpha=0.7, color='skyblue', edgecolor='black')
plt.hlines(theoretical_count[0], xmin=0.5, xmax=6.5, colors='red', linestyles='dashed', label='Theoretical Uniform')

plt.xlabel('Die Face')
plt.ylabel('Frequency')
plt.title('Dice Roll Distribution vs. Theoretical Uniform Distribution')
plt.xticks(values)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
plt.hist(dice_rolls, bins=np.arange(1, 8) - 0.5, edgecolor='black', rwidth=0.8)

# Step 3: Plot theoretical uniform line (expected count = 10000 / 6)
expected = len(dice_rolls) / 6
plt.axhline(expected, color='red', linestyle='--', label='Theoretical Count (≈1666.67)')

# Step 4: Customize plot
plt.xticks(range(1, 7))
plt.xlabel('Die Face')
plt.ylabel('Frequency')
plt.title('Distribution of 10,000 Dice Rolls vs. Theoretical Uniform Distribution')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Step 5: Show the plot
plt.tight_layout()
plt.show()
```
